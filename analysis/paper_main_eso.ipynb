```python
import xarray as xr
import numpy as np
import deepdish as dd

from codes import misc
import matplotlib.pyplot as plt
%matplotlib inline
```

# main figure 5

```python
data = xr.open_dataset(b'data/paper_eso.nc')
data = data.sel(delta_epsilon=-0.05, shape='square', grid=0.5)
data = misc.reduce_dimensions(data)
```

```python
labels = {
    'both': 'Total',
    'direct': 'Direct Rashba',
    'rashba': 'Rashba'
}
    

plt.figure(figsize=(3, 3))
for soi in data.soi:
    x = data.sel(soi=soi).Ex.data
    y = data.sel(soi=soi).energy_so  
    plt.plot(1000*x, 1000 * y, label=labels[str(soi.data)])

    
plt.xlabel(r'$E_x$ (V/$\mu$m)', fontsize=16)
plt.ylabel(r'$E_\mathrm{so} (meV)$', fontsize=16)
plt.xlim([0,10])
plt.tick_params(labelsize=15)
plt.tick_params(labelsize=15)
plt.legend(fontsize=12,frameon=False)


plt.savefig('figures/fig_5inset.pdf',transparent=True, bbox_inches='tight', pad_inches=0.1)
plt.savefig('figures/fig_5inset.png',transparent=True, bbox_inches='tight', pad_inches=0.1)
```
